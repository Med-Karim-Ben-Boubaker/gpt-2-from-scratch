```python
full_text = open("../../data/synthetic-data/3.txt", "r").read()
    
print(full_text[:1000])
```

```python
import re

# Split on word boundaries but keep spaces with following words
tokens = re.findall(r'\S+|\s+', full_text)
print(f"Tokens: {tokens[:100]}")

# Merge spaces with the following non-space tokens
results = []
i = 0
while i < len(tokens):
    if tokens[i].isspace() and i + 1 < len(tokens):
        # Merge space with the following word
        results.append(tokens[i] + tokens[i+1])
        i += 2
    else:
        results.append(tokens[i])
        i += 1
print(f"Updated Tokens: {results[:100]}")
```

```python
from collections import Counter

vocab_counter = Counter(results)
print(f"Total Vocab: {len(vocab_counter)}")
```

```python
# Print all the vocab with normalized frequency
for vocab, count in vocab_counter.most_common():
    print(f"{vocab}: {count / len(results)}")
```

```python
# Cumulative Probability analysis:
cumulative_prob = 0
token_count = 0
for vocab, count in vocab_counter.most_common():
    probability = count / len(results)
    cumulative_prob += probability
    token_count += 1
    
    # Print first 20 tokens for reference
    if token_count <= 20:
        print(f"{token_count:4d}. {vocab:15s}: {probability:.6f} (cumulative: {cumulative_prob:.6f})")
    
    # Stop when we reach 90%
    if cumulative_prob >= 0.95:
        break
    
print(f"Number of tokens to cover 90% of the text: {token_count}")
```

```python
import json
import tempfile
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.normalizers import Sequence, NFKC, Lowercase
from tokenizers.pre_tokenizers import WhitespaceSplit
from tokenizers.trainers import BpeTrainer

full_text = open("../../data/synthetic-data/3.txt", "r").read()

# Write your text to a temporary file
with tempfile.NamedTemporaryFile(mode='w', delete=False, suffix='.txt') as f:
    f.write(full_text)
    temp_file = f.name

normalizer = Sequence([NFKC(), Lowercase()])
tokenizer = Tokenizer(BPE(unk_token="<unk>", byte_fallback=True))
tokenizer.normalizer = normalizer
tokenizer.pre_tokenizer = WhitespaceSplit()

trainer = BpeTrainer(vocab_size=8000, show_progress=True, special_tokens=[])
tokenizer.train([temp_file], trainer)

tokenizer.save("./tokenizer.json")
```

```python
tokenizer = Tokenizer.from_file("./tokenizer.json")
print(f"Tokens: {tokenizer.encode(full_text).tokens[:100]}")
print(f"Token number: {len(tokenizer.encode(full_text).tokens)}")
```
